```python
import numpy as np
import matplotlib.pyplot as plt

from main import finite_element_method, estimate_error, y
```

```python
def draw(N, K):
    solution = [y(x) for x in np.linspace(0, 1, N)]
    plt.plot(np.linspace(0, 1, N), solution, label='real solution')
    plt.plot(np.linspace(0, 1, K + 1), finite_element_method(K), label='estimation')
    plt.title(f'h={1 / K}')
    plt.legend()
```

```python
draw(100000, 5)
```

```python
draw(100000, 20)
```

```python
draw(100000, 50)
```

```python
N_values = np.arange(10, 1000, 10)
h_square, max_error = estimate_error(N_values, 100000)
```

```python
plt.plot(N_values, h_square, label='h^2')
plt.plot(N_values, max_error, label='max error')
plt.title('h^2 vs max_error')
plt.legend()
```

```python
plt.plot(N_values, np.array(h_square) / np.array(max_error))
plt.title('h^2 / max_error')
```
